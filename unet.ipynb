## CNN

```python
# ======================================================================
# STEP 1: Imports
# ======================================================================
import os, cv2, numpy as np, matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import layers, models, backend as K
from tensorflow.keras.optimizers import Adam

# ======================================================================
# STEP 2: Load and preprocess BUSI dataset
# ======================================================================
DATA_ROOT = "/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT"
IMG_SIZE = 128
EPOCHS = 30
BATCH_SIZE = 8

images, masks = [], []
for cls in ["benign", "malignant"]:
    folder = os.path.join(DATA_ROOT, cls)
    all_files = sorted(os.listdir(folder))
    img_files = [f for f in all_files if f.endswith(".png") and "_mask" not in f]
    for img_file in img_files:
        img_path = os.path.join(folder, img_file)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE)).astype(np.float32) / 255.0

        # Combine multiple masks if present
        base = img_file.replace(".png", "")
        mask_files = [f for f in all_files if f.startswith(base + "_mask")]
        mask_combined = np.zeros((IMG_SIZE, IMG_SIZE), np.float32)
        for mfile in mask_files:
            mask = cv2.imread(os.path.join(folder, mfile), cv2.IMREAD_GRAYSCALE)
            mask = cv2.resize(mask, (IMG_SIZE, IMG_SIZE)).astype(np.float32) / 255.0
            mask_combined = np.maximum(mask_combined, mask)

        images.append(img)
        masks.append(mask_combined)

X = np.expand_dims(np.array(images), -1)
y = np.expand_dims(np.array(masks), -1)
print(f"âœ… Loaded {len(X)} imageâ€“mask pairs")
print("Shapes:", X.shape, y.shape)

X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1111, random_state=42)
print("Train:", X_train.shape, "Val:", X_val.shape, "Test:", X_test.shape)

# ======================================================================
# STEP 3: Define metrics and loss
# ======================================================================
def dice_coef(y_true, y_pred, smooth=1e-6):
    y_true_f, y_pred_f = K.flatten(y_true), K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2.*intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

def iou_coef(y_true, y_pred, smooth=1e-6):
    y_true_f, y_pred_f = K.flatten(y_true), K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    union = K.sum(y_true_f) + K.sum(y_pred_f) - intersection
    return (intersection + smooth) / (union + smooth)

def bce_dice_loss(y_true, y_pred):
    bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
    return bce + (1 - dice_coef(y_true, y_pred))

# ======================================================================
# STEP 4: CNN segmentation model (no skip connections)
# ======================================================================
def build_cnn_segmenter(input_shape=(IMG_SIZE, IMG_SIZE, 1)):
    inputs = layers.Input(shape=input_shape)

    # Feature extractor
    x = layers.Conv2D(32, (3,3), activation='relu', padding='same')(inputs)
    x = layers.Conv2D(32, (3,3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2,2))(x)

    x = layers.Conv2D(64, (3,3), activation='relu', padding='same')(x)
    x = layers.Conv2D(64, (3,3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2,2))(x)

    x = layers.Conv2D(128, (3,3), activation='relu', padding='same')(x)
    x = layers.Conv2D(128, (3,3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2,2))(x)

    x = layers.Conv2D(256, (3,3), activation='relu', padding='same')(x)
    x = layers.Conv2D(256, (3,3), activation='relu', padding='same')(x)

    # Upsampling to restore original resolution
    x = layers.UpSampling2D((2,2))(x)
    x = layers.Conv2D(128, (3,3), activation='relu', padding='same')(x)

    x = layers.UpSampling2D((2,2))(x)
    x = layers.Conv2D(64, (3,3), activation='relu', padding='same')(x)

    x = layers.UpSampling2D((2,2))(x)
    x = layers.Conv2D(32, (3,3), activation='relu', padding='same')(x)

    # Output mask
    outputs = layers.Conv2D(1, (1,1), activation='sigmoid')(x)
    model = models.Model(inputs, outputs, name="CNN_Segmenter")
    return model

model = build_cnn_segmenter()
model.compile(optimizer=Adam(1e-4), loss=bce_dice_loss, metrics=['accuracy', dice_coef, iou_coef])
model.summary()

# ======================================================================
# STEP 5: Train model
# ======================================================================
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    verbose=1
)

# ======================================================================
# STEP 6: Evaluate (Mean Dice & Mean IoU)
# ======================================================================
preds = model.predict(X_test, verbose=1)
preds_bin = (preds > 0.5).astype('float32')

def dice_np(y_true, y_pred):
    y_true, y_pred = y_true.flatten(), y_pred.flatten()
    inter = np.sum(y_true * y_pred)
    return (2*inter + 1e-6) / (np.sum(y_true) + np.sum(y_pred) + 1e-6)

def iou_np(y_true, y_pred):
    y_true, y_pred = y_true.flatten(), y_pred.flatten()
    inter = np.sum(y_true * y_pred)
    union = np.sum(y_true) + np.sum(y_pred) - inter
    return (inter + 1e-6) / (union + 1e-6)

dice_scores = [dice_np(y_test[i], preds_bin[i]) for i in range(len(y_test))]
iou_scores = [iou_np(y_test[i], preds_bin[i]) for i in range(len(y_test))]

print(f"\nğŸ“Š Mean Dice Coefficient: {np.mean(dice_scores):.4f}")
print(f"ğŸ“Š Mean IoU (Jaccard): {np.mean(iou_scores):.4f}")

# ======================================================================
# STEP 7: Visualization
# ======================================================================
n_show = 4
plt.figure(figsize=(12, n_show * 3))
for i in range(n_show):
    plt.subplot(n_show, 3, 3*i + 1)
    plt.imshow(X_test[i].squeeze(), cmap='gray')
    plt.title("Image"); plt.axis('off')

    plt.subplot(n_show, 3, 3*i + 2)
    plt.imshow(y_test[i].squeeze(), cmap='gray')
    plt.title("Ground Truth"); plt.axis('off')

    plt.subplot(n_show, 3, 3*i + 3)
    plt.imshow(preds_bin[i].squeeze(), cmap='gray')
    plt.title(f"Pred (Dice={dice_scores[i]:.2f}, IoU={iou_scores[i]:.2f})")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

## Dice

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
from sklearn.model_selection import train_test_split
from sklearn.metrics import jaccard_score, accuracy_score
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Conv2DTranspose, concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import backend as K


DATA_ROOT = "/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT"
IMG_SIZE = 128
EPOCHS = 30
BATCH_SIZE = 8

images, masks = [], []

for cls in ["benign", "malignant"]:
    folder = os.path.join(DATA_ROOT, cls)
    all_files = sorted(os.listdir(folder))

    # Get all image files (exclude masks)
    image_files = [f for f in all_files if f.endswith(".png") and "_mask" not in f]

    for img_file in image_files:
        img_path = os.path.join(folder, img_file)

        # Read and preprocess image
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = img.astype(np.float32) / 255.0

        # --- Combine all masks corresponding to this image ---
        base_name = img_file.replace(".png", "")
        mask_candidates = [
            f for f in all_files
            if f.startswith(base_name + "_mask") and f.endswith(".png")
        ]

        combined_mask = np.zeros((IMG_SIZE, IMG_SIZE), dtype=np.float32)

        for mask_file in mask_candidates:
            mask_path = os.path.join(folder, mask_file)
            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            mask = cv2.resize(mask, (IMG_SIZE, IMG_SIZE))
            mask = mask.astype(np.float32) / 255.0
            combined_mask = np.maximum(combined_mask, mask)  # combine all masks

        # Append the image and final combined mask
        images.append(img)
        masks.append(combined_mask)

# Convert to arrays
X = np.expand_dims(np.array(images), -1)
y = np.expand_dims(np.array(masks), -1)

print(f"âœ… Loaded {len(X)} combined imageâ€“mask pairs (benign + malignant)")
print("X shape:", X.shape, "| y shape:", y.shape)

X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1111, random_state=42)
print("Train:", X_train.shape, "Val:", X_val.shape,"Test:",X_test.shape)


def dice_coef(y_true, y_pred, smooth=1e-6):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

def dice_loss(y_true, y_pred):
    return 1 - dice_coef(y_true, y_pred)


def build_unet(input_shape=(IMG_SIZE, IMG_SIZE, 1)):
    inputs = Input(input_shape)
    
    # Encoder
    c1 = Conv2D(64, 3, activation='relu', padding='same')(inputs)
    c1 = Conv2D(64, 3, activation='relu', padding='same')(c1)
    p1 = MaxPooling2D((2,2))(c1)

    c2 = Conv2D(128, 3, activation='relu', padding='same')(p1)
    c2 = Conv2D(128, 3, activation='relu', padding='same')(c2)
    p2 = MaxPooling2D((2,2))(c2)

    c3 = Conv2D(256, 3, activation='relu', padding='same')(p2)
    c3 = Conv2D(256, 3, activation='relu', padding='same')(c3)
    p3 = MaxPooling2D((2,2))(c3)

    c4 = Conv2D(512, 3, activation='relu', padding='same')(p3)
    c4 = Conv2D(512, 3, activation='relu', padding='same')(c4)
    p4 = MaxPooling2D((2,2))(c4)

    # Bottleneck
    c5 = Conv2D(1024, 3, activation='relu', padding='same')(p4)
    c5 = Conv2D(1024, 3, activation='relu', padding='same')(c5)

    # Decoder
    u6 = Conv2DTranspose(512, 2, strides=2, padding='same')(c5)
    u6 = concatenate([u6, c4])
    c6 = Conv2D(512, 3, activation='relu', padding='same')(u6)
    c6 = Conv2D(512, 3, activation='relu', padding='same')(c6)

    u7 = Conv2DTranspose(256, 2, strides=2, padding='same')(c6)
    u7 = concatenate([u7, c3])
    c7 = Conv2D(256, 3, activation='relu', padding='same')(u7)
    c7 = Conv2D(256, 3, activation='relu', padding='same')(c7)

    u8 = Conv2DTranspose(128, 2, strides=2, padding='same')(c7)
    u8 = concatenate([u8, c2])
    c8 = Conv2D(128, 3, activation='relu', padding='same')(u8)
    c8 = Conv2D(128, 3, activation='relu', padding='same')(c8)

    u9 = Conv2DTranspose(64, 2, strides=2, padding='same')(c8)
    u9 = concatenate([u9, c1])
    c9 = Conv2D(64, 3, activation='relu', padding='same')(u9)
    c9 = Conv2D(64, 3, activation='relu', padding='same')(c9)

    outputs = Conv2D(1, 1, activation='sigmoid')(c9)

    model = Model(inputs, outputs)
    return model

model = build_unet()
model.compile(optimizer=Adam(1e-4),
              loss=dice_loss,
              metrics=['accuracy', dice_coef])

history = model.fit(X_train, y_train,
                    validation_data=(X_val, y_val),
                    epochs=50,
                    batch_size=BATCH_SIZE,
                    verbose=1)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score


preds = model.predict(X_test, batch_size=8, verbose=1)   # probabilities (0..1)
preds_bin = (preds > 0.5).astype(np.uint8)              # binary masks 0/1
y_true_bin = (y_test > 0.5).astype(np.uint8)


eps = 1e-6

def dice_np_single(y_true, y_pred):
    y_true_f = y_true.flatten()
    y_pred_f = y_pred.flatten()
    inter = (y_true_f * y_pred_f).sum()
    return (2.0 * inter + eps) / (y_true_f.sum() + y_pred_f.sum() + eps)

def iou_np_single(y_true, y_pred):
    y_true_f = y_true.flatten()
    y_pred_f = y_pred.flatten()
    inter = (y_true_f * y_pred_f).sum()
    union = y_true_f.sum() + y_pred_f.sum() - inter
    return (inter + eps) / (union + eps)

def acc_np_single(y_true, y_pred):
    return (y_true.flatten() == y_pred.flatten()).mean()

def precision_np_single(y_true, y_pred):
    y_true_f = y_true.flatten()
    y_pred_f = y_pred.flatten()
    TP = (y_true_f * y_pred_f).sum()
    FP = ((1 - y_true_f) * y_pred_f).sum()
    return (TP + eps) / (TP + FP + eps)

def recall_np_single(y_true, y_pred):
    y_true_f = y_true.flatten()
    y_pred_f = y_pred.flatten()
    TP = (y_true_f * y_pred_f).sum()
    FN = (y_true_f * (1 - y_pred_f)).sum()
    return (TP + eps) / (TP + FN + eps)

n = preds_bin.shape[0]
dice_scores = np.zeros(n, dtype=np.float32)
iou_scores  = np.zeros(n, dtype=np.float32)
acc_scores  = np.zeros(n, dtype=np.float32)
prec_scores = np.zeros(n, dtype=np.float32)
rec_scores  = np.zeros(n, dtype=np.float32)

for i in range(n):
    y_t = y_true_bin[i, ..., 0]  # H x W
    p   = preds_bin[i, ..., 0]   # H x W
    dice_scores[i] = dice_np_single(y_t, p)
    iou_scores[i]  = iou_np_single(y_t, p)
    acc_scores[i]  = acc_np_single(y_t, p)
    prec_scores[i] = precision_np_single(y_t, p)
    rec_scores[i]  = recall_np_single(y_t, p)

print("Dice Loss Results")
print(f"Mean Dice:      {dice_scores.mean():.4f}  |  Median: {np.median(dice_scores):.4f}")
print(f"Mean IoU:       {iou_scores.mean():.4f}  |  Median: {np.median(iou_scores):.4f}")
print(f"Mean Accuracy:  {acc_scores.mean():.4f}  |  Median: {np.median(acc_scores):.4f}")
print(f"Mean Precision: {prec_scores.mean():.4f}  |  Median: {np.median(prec_scores):.4f}")
print(f"Mean Recall:    {rec_scores.mean():.4f}  |  Median: {np.median(rec_scores):.4f}")
```

```python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# STEP 1: Imports
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
from sklearn.model_selection import train_test_split
from sklearn.metrics import jaccard_score, accuracy_score
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Conv2DTranspose, concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import backend as K


DATA_ROOT = "/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT"
IMG_SIZE = 128
EPOCHS = 30
BATCH_SIZE = 8

images, masks = [], []

for cls in ["benign", "malignant"]:
    folder = os.path.join(DATA_ROOT, cls)
    all_files = sorted(os.listdir(folder))

    # Get all image files (exclude masks)
    image_files = [f for f in all_files if f.endswith(".png") and "_mask" not in f]

    for img_file in image_files:
        img_path = os.path.join(folder, img_file)

        # Read and preprocess image
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = img.astype(np.float32) / 255.0

        # --- Combine all masks corresponding to this image ---
        base_name = img_file.replace(".png", "")
        mask_candidates = [
            f for f in all_files
            if f.startswith(base_name + "_mask") and f.endswith(".png")
        ]

        combined_mask = np.zeros((IMG_SIZE, IMG_SIZE), dtype=np.float32)

        for mask_file in mask_candidates:
            mask_path = os.path.join(folder, mask_file)
            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            mask = cv2.resize(mask, (IMG_SIZE, IMG_SIZE))
            mask = mask.astype(np.float32) / 255.0
            combined_mask = np.maximum(combined_mask, mask)  # combine all masks

        # Append the image and final combined mask
        images.append(img)
        masks.append(combined_mask)

# Convert to arrays
X = np.expand_dims(np.array(images), -1)
y = np.expand_dims(np.array(masks), -1)

print(f"âœ… Loaded {len(X)} combined imageâ€“mask pairs (benign + malignant)")
print("X shape:", X.shape, "| y shape:", y.shape)

X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1111, random_state=42)
print("Train:", X_train.shape, "Val:", X_val.shape,"Test:",X_test.shape)

# BCE + Dice hybrid loss
def bce_dice_loss(y_true, y_pred):
    bce = K.mean(K.binary_crossentropy(y_true, y_pred))
    dice = 1 - dice_coef(y_true, y_pred)
    return bce + dice

def tversky_loss(y_true, y_pred, alpha=0.7, beta=0.3, smooth=1e-6):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    TP = K.sum(y_true_f * y_pred_f)
    FP = K.sum((1-y_true_f) * y_pred_f)
    FN = K.sum(y_true_f * (1-y_pred_f))
    return 1 - (TP + smooth) / (TP + alpha*FP + beta*FN + smooth)


def build_unet(input_shape=(IMG_SIZE, IMG_SIZE, 1)):
    inputs = Input(input_shape)
    
    # Encoder
    c1 = Conv2D(64, 3, activation='relu', padding='same')(inputs)
    c1 = Conv2D(64, 3, activation='relu', padding='same')(c1)
    p1 = MaxPooling2D((2,2))(c1)

    c2 = Conv2D(128, 3, activation='relu', padding='same')(p1)
    c2 = Conv2D(128, 3, activation='relu', padding='same')(c2)
    p2 = MaxPooling2D((2,2))(c2)

    c3 = Conv2D(256, 3, activation='relu', padding='same')(p2)
    c3 = Conv2D(256, 3, activation='relu', padding='same')(c3)
    p3 = MaxPooling2D((2,2))(c3)

    c4 = Conv2D(512, 3, activation='relu', padding='same')(p3)
    c4 = Conv2D(512, 3, activation='relu', padding='same')(c4)
    p4 = MaxPooling2D((2,2))(c4)

    # Bottleneck
    c5 = Conv2D(1024, 3, activation='relu', padding='same')(p4)
    c5 = Conv2D(1024, 3, activation='relu', padding='same')(c5)

    # Decoder
    u6 = Conv2DTranspose(512, 2, strides=2, padding='same')(c5)
    u6 = concatenate([u6, c4])
    c6 = Conv2D(512, 3, activation='relu', padding='same')(u6)
    c6 = Conv2D(512, 3, activation='relu', padding='same')(c6)

    u7 = Conv2DTranspose(256, 2, strides=2, padding='same')(c6)
    u7 = concatenate([u7, c3])
    c7 = Conv2D(256, 3, activation='relu', padding='same')(u7)
    c7 = Conv2D(256, 3, activation='relu', padding='same')(c7)

    u8 = Conv2DTranspose(128, 2, strides=2, padding='same')(c7)
    u8 = concatenate([u8, c2])
    c8 = Conv2D(128, 3, activation='relu', padding='same')(u8)
    c8 = Conv2D(128, 3, activation='relu', padding='same')(c8)

    u9 = Conv2DTranspose(64, 2, strides=2, padding='same')(c8)
    u9 = concatenate([u9, c1])
    c9 = Conv2D(64, 3, activation='relu', padding='same')(u9)
    c9 = Conv2D(64, 3, activation='relu', padding='same')(c9)

    outputs = Conv2D(1, 1, activation='sigmoid')(c9)

    model = Model(inputs, outputs)
    return model

model = build_unet()
model.compile(optimizer=Adam(1e-4),
              loss=tversky_loss,
              metrics=['accuracy', dice_coef])

history = model.fit(X_train, y_train,
                    validation_data=(X_val, y_val),
                    epochs=50,
                    batch_size=BATCH_SIZE,
                    verbose=1)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score


preds = model.predict(X_test, batch_size=8, verbose=1)   # probabilities (0..1)
preds_bin = (preds > 0.5).astype(np.uint8)              # binary masks 0/1
y_true_bin = (y_test > 0.5).astype(np.uint8)


eps = 1e-6

def dice_np_single(y_true, y_pred):
    y_true_f = y_true.flatten()
    y_pred_f = y_pred.flatten()
    inter = (y_true_f * y_pred_f).sum()
    return (2.0 * inter + eps) / (y_true_f.sum() + y_pred_f.sum() + eps)

def iou_np_single(y_true, y_pred):
    y_true_f = y_true.flatten()
    y_pred_f = y_pred.flatten()
    inter = (y_true_f * y_pred_f).sum()
    union = y_true_f.sum() + y_pred_f.sum() - inter
    return (inter + eps) / (union + eps)

def acc_np_single(y_true, y_pred):
    return (y_true.flatten() == y_pred.flatten()).mean()

def precision_np_single(y_true, y_pred):
    y_true_f = y_true.flatten()
    y_pred_f = y_pred.flatten()
    TP = (y_true_f * y_pred_f).sum()
    FP = ((1 - y_true_f) * y_pred_f).sum()
    return (TP + eps) / (TP + FP + eps)

def recall_np_single(y_true, y_pred):
    y_true_f = y_true.flatten()
    y_pred_f = y_pred.flatten()
    TP = (y_true_f * y_pred_f).sum()
    FN = (y_true_f * (1 - y_pred_f)).sum()
    return (TP + eps) / (TP + FN + eps)

n = preds_bin.shape[0]
dice_scores = np.zeros(n, dtype=np.float32)
iou_scores  = np.zeros(n, dtype=np.float32)
acc_scores  = np.zeros(n, dtype=np.float32)
prec_scores = np.zeros(n, dtype=np.float32)
rec_scores  = np.zeros(n, dtype=np.float32)

for i in range(n):
    y_t = y_true_bin[i, ..., 0]  # H x W
    p   = preds_bin[i, ..., 0]   # H x W
    dice_scores[i] = dice_np_single(y_t, p)
    iou_scores[i]  = iou_np_single(y_t, p)
    acc_scores[i]  = acc_np_single(y_t, p)
    prec_scores[i] = precision_np_single(y_t, p)
    rec_scores[i]  = recall_np_single(y_t, p)

print("Trevsky Loss Results")
print(f"Mean Dice:      {dice_scores.mean():.4f}  |  Median: {np.median(dice_scores):.4f}")
print(f"Mean IoU:       {iou_scores.mean():.4f}  |  Median: {np.median(iou_scores):.4f}")
print(f"Mean Accuracy:  {acc_scores.mean():.4f}  |  Median: {np.median(acc_scores):.4f}")
print(f"Mean Precision: {prec_scores.mean():.4f}  |  Median: {np.median(prec_scores):.4f}")
print(f"Mean Recall:    {rec_scores.mean():.4f}  |  Median: {np.median(rec_scores):.4f}")
```
